# Data Cleaning

```python
# total number of duplicates in a DataFrame(DF)
df.duplicated().sum()
```

```python
# removing outliers in SepalLengthCm column by using IQR Method
Q1 = df['SepalLengthCm'].quantile(0.25)
Q3 = df['SepalLengthCm'].quantile(0.75)
IQR = Q3 - Q1
bmi_lower_bound = Q1 - 1.5*IQR
bmi_upper_bound = Q3 + 1.5*IQR
df = df[(df['SepalLengthCm'] > bmi_lower_bound)&(df['SepalLengthCm']<bmi_upper_bound)]
print(df.shape)
```

```python
# removing outliers in SepalWidthCm column by using IQR Method
Q1 = df['SepalWidthCm'].quantile(0.25)
Q3 = df['SepalWidthCm'].quantile(0.75)
IQR = Q3 - Q1
bmi_lower_bound = Q1 - 1.5*IQR
bmi_upper_bound = Q3 + 1.5*IQR
df = df[(df['SepalWidthCm'] > bmi_lower_bound)&(df['SepalWidthCm']<bmi_upper_bound)]
print(df.shape)
```

```python
# removing outliers in PetalLengthCm column by using IQR Method
Q1 = df['PetalLengthCm'].quantile(0.25)
Q3 = df['PetalLengthCm'].quantile(0.75)
IQR = Q3 - Q1
bmi_lower_bound = Q1 - 1.5*IQR
bmi_upper_bound = Q3 + 1.5*IQR
df = df[(df['PetalLengthCm'] > bmi_lower_bound)&(df['PetalLengthCm']<bmi_upper_bound)]
print(df.shape)
```

```python
# removing outliers in  PetalWidthCm column by using IQR Method
Q1 = df['PetalWidthCm'].quantile(0.25)
Q3 = df['PetalWidthCm'].quantile(0.75)
IQR = Q3 - Q1
bmi_lower_bound = Q1 - 1.5*IQR
bmi_upper_bound = Q3 + 1.5*IQR
df = df[(df['PetalWidthCm'] > bmi_lower_bound)&(df['PetalWidthCm']<bmi_upper_bound)]
print(df.shape)
```

# Identifying and segregating the Input Variables (x) and Output Variables(y):

```python
y = df['Species']
X = df.drop("Species",axis=1)
```

# Splitting the data into Training Data and Testing Data:

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

# Data Preparation on Training Data:

# Applying Standardization Technique on Training Data:

```python
from sklearn.preprocessing import StandardScaler
StandardScaler_object = StandardScaler()
X_Train_Transformed = pd.DataFrame(StandardScaler_object.fit_transform(X_train), columns = X_train.columns, index = X_train.index)
print(X_Train_Transformed.shape)
```

# Building a Model by using KNN classifier Algorithm:

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=5, p=2, metric='minkowski')

classifier.fit(X_Train_Transformed, y_train)
```

```python
# Classes known to classifier
classifier.classes_
```

```python
# Knowing the distance metric used
classifier.effective_metric_
```

```python
# Number of features seen by algo during training
classifier.n_features_in_
```

```python
# Names of features seen during training
classifier.feature_names_in_
```

# Data Preparation on Test Data:

```python
X_Test_Transformed = pd.DataFrame(StandardScaler_object.transform(X_test),
                                 columns = X_test.columns,
                                 index = X_test.index)

print(X_Test_Transformed.shape)
```

# Predictions:

```python
y_test_pred = classifier.predict(X_Test_Transformed)
```

# Evaluation of a Model:

```python
# calculate accuracy of class predictions
from sklearn import metrics

metrics.accuracy_score(y_test, y_test_pred)
```

# Training the Model with Cosine Distance Metric:

```python
from sklearn.neighbors import KNeighborsClassifier

KNNclassifier_object = KNeighborsClassifier(metric='cosine')

KNNclassifier_object.fit(X_Train_Transformed, y_train)
```

```python
# Knowing the distance metric used
KNNclassifier_object.effective_metric_
```

```python
X_test_transformed = pd.DataFrame(StandardScaler_object.transform(X_test),
                                 columns = X_test.columns,
                                 index = X_test.index)

print(X_test_transformed.shape)
```

# Predictions:

```python
y_test_pred = KNNclassifier_object.predict(X_test_transformed)
```

# Evaluation of a Model:

```python
# calculate accuracy of class predictions
from sklearn import metrics

metrics.accuracy_score(y_test, y_test_pred)
```
