```python
# removing outliers in SepalLengthCm column by using IQR Method
Q1 = df['SepalLengthCm'].quantile(0.25)
Q3 = df['SepalLengthCm'].quantile(0.75)
IQR = Q3 - Q1
bmi_lower_bound = Q1 - 1.5*IQR
bmi_upper_bound = Q3 + 1.5*IQR
df = df[(df['SepalLengthCm'] > bmi_lower_bound)&(df['SepalLengthCm']<bmi_upper_bound)]
print(df.shape)
```

```python
# removing outliers in SepalWidthCm column by using IQR Method
Q1 = df['SepalWidthCm'].quantile(0.25)
Q3 = df['SepalWidthCm'].quantile(0.75)
IQR = Q3 - Q1
bmi_lower_bound = Q1 - 1.5*IQR
bmi_upper_bound = Q3 + 1.5*IQR
df = df[(df['SepalWidthCm'] > bmi_lower_bound)&(df['SepalWidthCm']<bmi_upper_bound)]
print(df.shape)
```

```python
# removing outliers in PetalLengthCm column by using IQR Method
Q1 = df['PetalLengthCm'].quantile(0.25)
Q3 = df['PetalLengthCm'].quantile(0.75)
IQR = Q3 - Q1
bmi_lower_bound = Q1 - 1.5*IQR
bmi_upper_bound = Q3 + 1.5*IQR
df = df[(df['PetalLengthCm'] > bmi_lower_bound)&(df['PetalLengthCm']<bmi_upper_bound)]
print(df.shape)
```

```python
# removing outliers in  PetalWidthCm column by using IQR Method
Q1 = df['PetalWidthCm'].quantile(0.25)
Q3 = df['PetalWidthCm'].quantile(0.75)
IQR = Q3 - Q1
bmi_lower_bound = Q1 - 1.5*IQR
bmi_upper_bound = Q3 + 1.5*IQR
df = df[(df['PetalWidthCm'] > bmi_lower_bound)&(df['PetalWidthCm']<bmi_upper_bound)]
print(df.shape)
```
